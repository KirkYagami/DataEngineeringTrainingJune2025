## Basics of Matplotlib

### What is Matplotlib
The [matplotlib](http://matplotlib.org) library is a powerful tool capable of producing complex publication-quality figures with fine layout control in two and three dimensions. While it is an older library, so many libraries are built on top of it and use its syntax.

It is widely used for data visualization in data science and analytics

```python
import matplotlib.pyplot as plt
```

```python

# where do we use line plot? - time series analysis, etc..

x=[1,2,3,4,5]
y=[1,4,9,16,25]

# y = x^2

plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel(" Y Axis")
plt.title("basic line plot for y = x^2")
plt.show()
```

```python

x=[1,2,3,4,5]
y=[1,8,27,64,125]

# y = x^2

plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel(" Y Axis")
plt.title("basic line plot for y = x^3")
plt.show()
```

```python
# Customization

x=[1,2,3,4,5]
y=[1,8,27,64,125]

plt.figure(figsize=(7,3))
plt.plot(x, y,
         color='b',
         linestyle="--",
         marker='h',
         linewidth=3,
         markersize=12
         )

plt.grid(True)

plt.show()
```

```python
# subplots

x=[1,2,3,4,5]
y1 = [1,4,9,16,25]
y2=[1,8,27,64,125]


plt.figure(figsize=(12,8))

plt.subplot(2,2,1)


plt.plot(x, y1,
         color='b',
         linestyle="--",
         marker='h',
         linewidth=3,
         markersize=12
         )
plt.title('Plot 1')

plt.subplot(2,2,2)


plt.plot(x, y2,
         color='r',
         linestyle="--",
         marker='h',
         linewidth=3,
         markersize=12
         )
plt.title('Plot 2')


plt.subplot(2,2,3)


plt.plot(x, y2,
         color='g',
         linestyle="--",
         marker='h',
         linewidth=3,
         markersize=12
         )
plt.title('Plot 3')




plt.subplot(2,2,4)


plt.plot(x, y1,
         color='black',
         linestyle="--",
         marker='o',
         linewidth=3,
         markersize=12
         )
plt.title('Plot 2')


plt.grid(True)

plt.show()
```

```python
# bar

cat = ["A", "B", "C", "D", "E"]

values = [10, 7, 15, 4, 9]


plt.bar(cat, values, color='purple')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Sample Bar Chart")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,3,4,5,6]

plt.scatter(x,y, color='b', marker='x')
plt.show()
```

```python
cat = ["A", "B", "C", "D"]
sizes = [30,20,40,10]

colors = ['gold', "forestgreen", "blue", "lightskyblue"]

explode = (0.2, 0, 0, 0)
plt.pie(sizes, labels=cat,
         explode=explode,
         colors=colors,
         autopct="%1.1f%%",
         shadow=True)

plt.savefig("pie_chart.png", dpi=300)
plt.show()
```
