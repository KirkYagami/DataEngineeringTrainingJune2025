1. **Equality Operator (`'=='`):**
   - Checks if two values are equal.
   ```python
   x = 5
   y = 5
   result = x == y  # True
   ```

2. **Inequality Operator (`!=`):**
   - Checks if two values are not equal.
   ```python
   x = 5
   y = 10
   result = x != y  # True
   ```

3. **Less Than Operator (`<`):**
   - Checks if the value on the left is less than the value on the right.
   ```python
   x = 5
   y = 10
   result = x < y  # True
   ```

4. **Greater Than Operator (`>`):**
   - Checks if the value on the left is greater than the value on the right.
   ```python
   x = 15
   y = 10
   result = x > y  # True
   ```

5. **Less Than or Equal To Operator (`<=`):**
   - Checks if the value on the left is less than or equal to the value on the right.
   ```python
   x = 10
   y = 10
   result = x <= y  # True
   ```

6. **Greater Than or Equal To Operator (`>=`):**
   - Checks if the value on the left is greater than or equal to the value on the right.
   ```python
   x = 20
   y = 15
   result = x >= y  # True
   ```

7. **Membership Operators (`in`, `not in`):**
   - `in` checks if a value exists in a sequence.
   - `not in` checks if a value does not exist in a sequence.
   ```python
   my_list = [1, 2, 3, 4, 5]
   result1 = 3 in my_list  # True
   result2 = 'a' not in 'hello'  # True
   ```

8. **Identity Operators (`is`, `is not`):**
   - `is` checks if two variables refer to the same object.
   - `is not` checks if two variables do not refer to the same object.
   ```python
   x = [1, 2, 3]
   y = [1, 2, 3]
   result1 = x is y  # False
   result2 = x is not y  # True
   ```


````python
my_list = [1, 2, 3, 4, 5, 3.14,'nick']
result1 = 3 in my_list  # True
result2 = 'a' not in 'hello'  # True

print(result1)
print(result2)
````
